```python
import pandas as pd
%matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=pd.read_csv("data.csv")
```

```python
montreal=data[data["team.name"]=="Montréal Canadiens"].tail(500)
def get_sec(time_str):
    m, s = time_str.split(':')
    return 3600 + int(m) * 60 + int(s)
second=[]
for time in montreal["about.periodTime"]:
    second.append(get_sec(time))
montreal['TimeinSecond']=second
```

```python
montreal
```

```python
def positionWithTime(time=20):
    filtered=montreal.loc[(montreal['TimeinSecond'] >= time-20) & (montreal['TimeinSecond'] <= time)]
    fix,ax=plt.subplots(figsize=(16,10))
    filtered.plot(kind='scatter',
                         x='coordinates.x',
                         y='coordinates.y',
                         ax=ax);
```

```python
import ipywidgets as widgets
widgets.interact(positionWithTime,time=(20,1200))
```

```python

slider = widgets.IntSlider(value=50, min=montreal["TimeinSecond"].min(), max=montreal["TimeinSecond"].max(), continuous_update=False)
slider.style.handle_color="green"
slider
```
