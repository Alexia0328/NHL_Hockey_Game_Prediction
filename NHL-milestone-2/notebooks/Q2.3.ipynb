```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np
import math
```

```python
df_dataset_R = pd.read_csv("Milestone2_newData.csv")
goal = df_dataset_R[df_dataset_R['shot or goal'] == 'Goal'] 
nonGoal = df_dataset_R[df_dataset_R['shot or goal'] == 'Shot']
goalAngle=goal["Angle from net"].dropna()
nonGoalAngle=nonGoal["Angle from net"].dropna()
goalAngle
```

```python
sumGoal=list()
sum=list()
for x in range(360):
    sumGoal.append(0)
    sum.append(0)
for val in goalAngle:
    sum[math.trunc(val)]+=1
    sumGoal[math.trunc(val)]+=1
for val in nonGoalAngle:
    sum[math.trunc(val)]+=1
```

```python
angleRate=list()
for x in range(360):
    angleRate.append(0)
for idx, val in enumerate(sumGoal):
    if(sum[idx]):
        angleRate[idx]=val/sum[idx]
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
plt.xlabel('Angle of the goal')
plt.ylabel('Goal rate')
ax.plot(range(360),angleRate)
plt.show()
```
