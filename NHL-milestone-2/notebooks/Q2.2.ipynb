```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np
import math
```

```python
df_dataset_R = pd.read_csv("Milestone2_newData.csv")
goal = df_dataset_R[df_dataset_R['shot or goal'] == 'Goal'] 
nonGoal = df_dataset_R[df_dataset_R['shot or goal'] == 'Shot']
goalDistance=goal["Distance from net"].dropna()
nonGoalDistance=nonGoal["Distance from net"].dropna()
goalDistance
```

```python
sumGoal=list()
sum=list()
for x in range(200):
    sumGoal.append(0)
    sum.append(0)
for val in goalDistance:
    sum[math.trunc(val)]+=1
    sumGoal[math.trunc(val)]+=1
for val in nonGoalDistance:
    sum[math.trunc(val)]+=1
```

```python
distanceRate=list()
for x in range(200):
    distanceRate.append(0)
for idx, val in enumerate(sumGoal):
    if(sum[idx]):
        distanceRate[idx]=val/sum[idx]
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
plt.xlabel('Distance of the goal')
plt.ylabel('Goal rate')
ax.plot(range(200),distanceRate)
```
